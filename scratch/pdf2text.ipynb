```python
import pandas as pd
import pytesseract
from pdf2image import convert_from_path
from PIL import Image, ImageDraw

from myrsidea.pylib.const import PDF_DIR, TXT_DIR, DATA_DIR
from myrsidea.pylib.util import make_temp_dir
```

```python
TEMP_DIR = DATA_DIR / 'temp'
PDF = PDF_DIR / '68930.pdf'
```

```python
COLORS = {1: 'gray', 2: 'red', 3: 'blue', 4: 'green', 5: 'purple'}
```

```python
with make_temp_dir(TEMP_DIR, prefix='pages_', keep=True) as temp_dir:
    images = convert_from_path(PDF, output_folder=temp_dir)
    for i, image in enumerate(images, 1):
        draw = ImageDraw.Draw(image)
        text = pytesseract.image_to_string(image)
        print(text)
        print('=' * 80)
        df = pytesseract.image_to_data(image, output_type='data.frame')
        for idx, row in df.iterrows():
            if row.level <= 5:
                color = COLORS[row.level]

                pad = (5 - row.level) * 2
                left = row.left - pad
                top = row.top - pad
                right = row.left + row.width - 1 + pad
                bottom = row.top + row.height - 1 + pad

                bbox = (left, top, right, bottom)
                draw.rectangle(bbox, outline=color)
#         image.save(temp_dir + f'/page_{i}.jpg', 'JPEG')
```
